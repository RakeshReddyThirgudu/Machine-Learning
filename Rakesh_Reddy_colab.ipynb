```
#convert the file into a dataframe and take a look at first couple rows
import pandas as pd
dataset = pd.read_csv('/content/drive/MyDrive/MIS/50_Startups.csv')
print(dataset.head(3))
```

```
#summary statistics for the dataset
dataset.describe()
```

```
# data visualization with seaborn library
import seaborn as sns
sns.distplot(dataset['R&D Spend']) #histogram of the "R&D spend" variable
```

```
sns.scatterplot(x="R&D Spend",y="Marketing Spend", hue='State', data=dataset)
```

```
# plot the dataset with altair library (interactive), referencing dataframe column names (change column names to explore)
import altair as alt
alt.Chart(dataset).mark_point().encode(
  x='R&D Spend',
  y='Marketing Spend',
  color='State'
  ).interactive()
```
